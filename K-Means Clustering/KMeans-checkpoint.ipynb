```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("income.csv")
```

```python
data
```

```python
%matplotlib inline
plt.figure(figsize=(12,6))
plt.scatter(x=data.Age,y=data['Income($)'])
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3)
km
```

```python
X=data.drop(['Income($)','Name'],axis=1)
```

```python
X
```

```python
y=data['Income($)']
```

```python
y
```

```python
pred=km.fit_predict(X,y)
pred
```

```python
data['Cluster1']=pred
```

```python
data
```

```python
data0 = data[data.Cluster1==0]
data1 = data[data.Cluster1==1]
data2 = data[data.Cluster1==2]
plt.scatter(x=data0.Age,y=data0['Income($)'],color='Green')
plt.scatter(x=data1.Age,y=data1['Income($)'],color='Black')
plt.scatter(x=data2.Age,y=data2['Income($)'],color='Red')

plt.legend()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(data[['Income($)']])
data['Income($)']=scaler.transform(data[['Income($)']])

scaler.fit(data[['Age']])
data['Age']=scaler.transform(data[['Age']])

data
```

```python
km = KMeans(n_clusters=3)
km
```

```python
X = data.drop(['Income($)','Name','Cluster1','Cluster2'],axis=1)
y= data['Income($)']
X
```

```python
y
```

```python
pred2=km.fit_predict(data[['Age','Income($)']])
pred2
```

```python
data['Cluster2']=pred2
data
```

```python
dataa = data[data.Cluster2==0]
datab = data[data.Cluster2==1]
datac = data[data.Cluster2==2]
plt.scatter(x=dataa.Age,y=dataa['Income($)'],color='Green')
plt.scatter(x=datab.Age,y=datab['Income($)'],color='Black')
plt.scatter(x=datac.Age,y=datac['Income($)'],color='Red')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='yellow',marker='*',label='centroid')

plt.legend()
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(data[['Age','Income($)']])
    sse.append(km.inertia_)
```

```python
plt.plot(k_rng,sse)
```

#### Elbow technique, the elbow of the graph suggests that k=3
