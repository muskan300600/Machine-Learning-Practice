```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('price.csv')
```

```python
data
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x=data.area,y=data.price)
```

```python
plt.figure(figsize=(12,6))
sns.regplot(x=data.area,y=data.price,color='red',marker='*')
```

```python
X = data.drop('price',axis=1)
```

```python
X
```

```python
y = data.price
```

```python
y
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
```

```python
lr.predict([[3000]])
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
area = pd.read_csv('area.csv')
```

```python
area
```

```python
area['Predicted Price'] = lr.predict(area)
```
