```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Melbourne_housing_FULL.csv')
data
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data.nunique()
```

```python
cols=['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom'
           ,'Car','Landsize','BuildingArea','Price']
data=data[cols]
data
```

```python
data.isnull().sum()
```

```python
cols_zero = ['Propertycount','Distance','Bedroom2','Bathroom','Car']
data[cols_zero]=data[cols_zero].fillna(0)
data.isnull().sum()
```

```python
data.Landsize=data.Landsize.fillna(data.Landsize.mean())
data.BuildingArea=data.BuildingArea.fillna(data.BuildingArea.mean())
data.isnull().sum()
```

```python
data.dropna(inplace=True)
data.isnull().sum()
```

```python
data.info()
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Suburb']=le.fit_transform(data.Suburb)
data.Type = le.fit_transform(data.Type)
data.Method = le.fit_transform(data.Method)
data.SellerG=le.fit_transform(data.SellerG)
data.Regionname=le.fit_transform(data.Regionname)
data.CouncilArea = le.fit_transform(data.CouncilArea)
data
```

```python
data.info()
```

```python
X = data.drop(['Price'],axis=1)
y=data.Price
```

```python
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
X_train.shape,X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
from sklearn import metrics
pred = lr.predict(X_test)
score = metrics.r2_score(pred,y_test)
score
```

```python
lr.score(X_test,y_test)
```

```python
lr.score(X_train,y_train)
```

L1 regularization or Lasso Model

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=5,max_iter=100,tol=0.1)
lasso_reg.fit(X_train,y_train)
```

```python
pred=lasso_reg.predict(X_test)
score=metrics.r2_score(pred,y_test)
score
```

```python
lasso_reg.score(X_train,y_train)
```

```python
lasso_reg.score(X_test,y_test)
```

```python
from sklearn import linear_model
ridge_reg = linear_model.Ridge(alpha=5,max_iter=100,tol=0.1)
ridge_reg.fit(X_train,y_train)
```

```python
pred=ridge_reg.predict(X_test)
score=metrics.r2_score(pred,y_test)
score
```

```python
ridge_reg.score(X_train,y_train)
```

```python
ridge_reg.score(X_test,y_test)
```
